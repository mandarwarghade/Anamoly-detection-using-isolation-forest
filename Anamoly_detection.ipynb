```python
import matplotlib.pyplot as plt
import numpy as np
import os
from datetime import datetime 
import pandas as pd
import plotly.express as px
from sklearn.ensemble import IsolationForest

plt.figure(figsize=(13,8))
```

```python
df=pd.read_csv("nyc_taxi.csv")
df.head()
```

```python
df.dtypes
```

```python
df['timestamp']=pd.to_datetime(df['timestamp'])
```

```python
df.dtypes
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df=df.set_index('timestamp').resample('H').mean().reset_index()
```

```python
df.head()
```

```python
df['hour']=df.timestamp.dt.hour
```

```python
df['weekday']=pd.Categorical(df.timestamp.dt.strftime('%A'),categories=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday',"Sunday"],ordered=True)
df
```

```python
df[['value','weekday']].groupby('weekday').mean().plot()
plt.show()
```

```python
df[['value','hour']].groupby('hour').mean().plot()
plt.show()
```

```python
fig=px.line(df,x=df['timestamp'],y=df['value'],title='NYC Taxi Demand')

fig.update_xaxes(
    rangeslider_visible=True,
)
fig.show()
```

```python
model=IsolationForest(contamination=0.004)
model.fit(df[['value']])
```

```python
df['outliers']=pd.Series(model.predict(df[['value']])).apply(lambda x:'yes' if(x==-1) else 'no')
```

```python
df.query('outliers=="yes"')
```

```python
fig=px.scatter(df,x=df['timestamp'],y=df['value'],color='outliers',hover_data=['weekday'],title='NYC Taxi Demand')

fig.update_xaxes(
    rangeslider_visible=True,
)

fig.show()
```
